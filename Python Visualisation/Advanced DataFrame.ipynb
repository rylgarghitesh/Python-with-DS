# Seaborn Gallery 

```python
sns.lmplot(data = stats, 
           x = "Internet users", 
           y = "Birth rate",
          fit_reg=False)
```

```python
sns.lmplot(data = stats, 
           x = "Internet users", 
           y = "Birth rate",
          fit_reg=False,
          hue="Income Group", aspect=1.5)
```

```python
sns.lmplot(data = stats, 
           x = "Internet users", 
           y = "Birth rate",
          fit_reg=False,
          hue="Income Group", 
           aspect=1.5,
           scatter_kws= {"s":100}
          )
```
